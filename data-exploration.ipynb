```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax=plt.axes()
sns.heatmap(books.isna().transpose(),cbar=False,ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.show()
```

```python
import numpy as np

books["missing_description"]= np.where(books["description"].isna(),1,0)
books["age_of_book"]=2025-books["published_year"]
```

```python
columns_of_interest=["num_pages","age_of_book","missing_description","average_rating"]

correlation_matirx=books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap=sns.heatmap(correlation_matirx,annot=True,fmt=".2f",cmap="coolwarm",cbar_kws={"label":"Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

This heatmap visualizes the Spearman correlation between different features in the dataset:

num_pages: Number of pages in a book

age_of_book: How old the book is

missing_description: Whether a book is missing a description (1 = missing, 0 = present)

average_rating: Average rating of the book

📘 Key Insights:
The correlation values range from -1 (perfect negative) to +1 (perfect positive). Values near 0 indicate no meaningful relationship.

The strongest correlation here is between num_pages and average_rating (+0.15), suggesting that longer books tend to get slightly higher ratings — though the relationship is weak.

age_of_book has a small positive correlation with missing_description (+0.13), which means older books are slightly more likely to have missing descriptions.

average_rating has very weak or no correlation with missing_description or age_of_book, meaning that factors like book age or missing descriptions don’t significantly impact ratings.

Most variables in this dataset show very weak correlations, indicating that these features may not have a strong linear or monotonic influence on each other.

```python
books[(books["description"].isna()) | (books["num_pages"].isna()) | (books["average_rating"].isna()) | (books["published_year"].isna())]
```

```python
books_missing=books[~(books["description"].isna()) & ~(books["num_pages"].isna()) & ~(books["average_rating"].isna()) & ~(books["published_year"].isna())]
```

```python
books_missing
```

Filtered out rows that had missing values in crucial columns: description, num_pages, average_rating, and published_year. A total of 303 incomplete entries were removed. The cleaned dataset, stored in books_missing, contains 6507 complete records with all necessary fields present — making it suitable for building a high-quality recommender system.

```python
books_missing['categories'].value_counts().reset_index().sort_values("count",ascending=False)

# This line helps us understand which book categories (genres/topics) are most common in our cleaned dataset.

# value_counts() counts how many times each category appears in the 'categories' column.
# reset_index() turns this count series into a structured DataFrame for easier viewing.
# sort_values("count", ascending=False) arranges the categories from most to least frequent.

# This step helps identify the most common book categories in the dataset, which can guide genre-based recommendations and improve user personalization.
```

```python
# Plot a bar chart to visualize how frequently each book category appears in the dataset
books_missing['categories'].value_counts().head(20).plot(kind='bar', figsize=(12,6), title='Top 20 Book Categories by Frequency', xlabel='Category', ylabel='Count')
```

```python
category_counts = books_missing['categories'].value_counts().head(100)
category_counts.plot(kind='bar', figsize=(20,10), title='Top 100 Book Categories by Frequency')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

```python
books_missing["words_in_description"]=books_missing["description"].str.split().str.len()
```

```python
books_missing
```

```python
# Remove rows with negative word counts
books_missing = books_missing[books_missing["words_in_description"] >= 0]
```

```python
books_missing
```

```python
import plotly.express as px

# Plot histogram of description lengths
fig = px.histogram(books_missing, 
                   x="words_in_description", 
                   nbins=50, 
                   title="Distribution of Words in Book Descriptions", 
                   color_discrete_sequence=["indianred"])

fig.update_layout(xaxis_title="Words in Description",
                  yaxis_title="Count")

fig.show()
```

```python
books_missing_25_words = books_missing[books_missing["words_in_description"] >= 25]
```

```python
books_missing_25_words
```

```python
books_missing_25_words["title_and_subtitle"]=(
    np.where(books_missing_25_words["subtitle"].isna(), books_missing_25_words["title"],
             books_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join,axis=1))
             )
```

```python
books_missing_25_words
```

```python
books_missing_25_words["tagged_description"]=books_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
```
